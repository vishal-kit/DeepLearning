```python
# import modules

import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook

scores = [3.0, 1.0, 0.2]
```

```python
def softmax(x):
    """Compute the softmax valuses for X"""
    #x= np.asarray(x)
    return np.exp(x)/np.sum(np.exp(x),axis=0) 

print (softmax(scores))
```

```python
x= np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2*np.ones_like(x)])
```

```python
# Plot softmax curves
plt.plot(x,softmax(scores).T,linewidth=2)
plt.legend(['x','1','0.2'])
```

```python
plt.plot(x, softmax(scores).T, linewidth=2)
plt.show()
```
